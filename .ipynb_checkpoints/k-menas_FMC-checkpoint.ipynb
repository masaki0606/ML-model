## 2. k_means法の仕組みについて

- あらかじめクラスター数(k)を決めておいて、その数だけデータ全体を分割するのが、k-means法
- クラスタ内にデータが存在する時に、各データ点とセントロイド(重心)の距離の差の２乗(SSE)を総和したものを最小にするという最適化問題
- SSEはwに対しての一次式で表される(制約条件もuについて一次方程式であるため最適解は必ず直線の端点の0か1になる)(ノルムの中身は実数値)

$
min   \sum_{i=1}^{n}w^{(i,j)}\begin{Vmatrix}x^{(i)}-μ^{(j)}\\ \end{Vmatrix}^2_2
$

$
s.t. w^{(i,j)} \in \{0,1\}
$

<b>【プロトタイプについて】</b>

| プロトタイプ 　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　                | 概要                                                                                         | 
| --------------------------------------------------------| ---------------------------------------------------------------------------------------------- :| 
| セントロイド                                           | 特徴量が連続値の時、類似する点の重心を表す                                              | 
| メドイド                                                | 特徴量がカテゴリのとき、最も代表的な点を表す |

<b>0. 準備2 -kmeansを実装- (参考：https://github.com/hhachiya/MLBook/blob/master/codes/kmeans.py)</b>

```python
class kmeans:
    #-------------------
    #  k-meansの各種初期化
    # X: 学習データ（データ数×次元数のnumpy.ndarray）
    # K: クラスター数（整数スカラー）
    def __init__(self,X,K=5):
        # パラメータの設定
        self.dNum = len(X)
        self.K = K
        self.X = X
        
        # カラーコードの設定
        self.cmap = ['#FF0000','#00B0F0','#FF00FF','#00FF00','#0000FF']

        # STEP1_ランダムにクラスター中心(セントロイド)を設定
        # np.random.permutation : 配列の要素をランダムに並び替えて、新しい配列を生成
        self.C = X[np.random.permutation(self.dNum)[:self.K],:]
    #-------------------
    #-------------------
    # クラスターの中心の更新をするメソッド
    def updateCluster(self):
        
        # STEP2_XとCの全ペア間の距離の計算
        #np.tile : n次元配列を繰り返す
        #np.expand_dims : 次元を一つ増やす
        Ctmp = np.tile(np.expand_dims(self.C.T,axis=2),[1,1,self.dNum])
        Xtmp = np.tile(np.expand_dims(self.X.T,axis=1),[1,self.K,1])
        dist = np.sum(np.square(Ctmp - Xtmp),axis=0)
        
        # STEP2_距離が最小のクラスターのインデックスを選択
        self.cInd = np.argmin(dist,axis=0)
        
        # STEP3_各クラスタに属しているデータ点の平均値(重心)を計算し、新しいクラスター中心に設定
        self.C = np.array([np.mean(self.X[self.cInd==c],axis=0) for c in range(self.K)])
    #-------------------
    
    #-------------------
    # 3. 学習データとクラスターのプロット
    # （特徴数が2の場合）
    # fName：画像の保存先（文字列）
    def plotCluster(self,xLabel="$x_1$",yLabel="$x_2$",fName=""):
        plt.close()
        
        # クラスターごとに学習データとクラスター中心のプロット
        for c in range(self.K):
            plt.plot(self.X[self.cInd==c,0],self.X[self.cInd==c,1],'s',color=self.cmap[c],markeredgecolor='k',markersize='8')
            plt.plot(self.C[c,0],self.C[c,1],'o',color=self.cmap[c],markeredgecolor='k',markersize='16')

        # ラベル、グリッドおよび範囲の設定
        plt.xlabel(xLabel,fontsize=14)
        plt.ylabel(yLabel,fontsize=14)
        plt.tick_params(labelsize=14)
        plt.grid(True)
        plt.xlim(np.min(self.X[:,0]),np.max(self.X[:,0]))
        plt.ylim(np.min(self.X[:,1]),np.max(self.X[:,1]))


        # グラフの表示またはファイルへの保存
        if len(fName):
            plt.savefig(fName)
        else:
            plt.show()
```


STEP1_クラスタ中心の初期値として、データ点からK個（今回は2個）のセントロイドをランダムに選ぶ  
STEP2_各データ点を最も近いセントロイドに割り当てる  
STEP3_セントロイドに割り当てられたデータ点の中心に、セントロイドを移動する  
STEP4_データ点へのクラスタの割り当てが変化しなくなるか、ユーザー定義の許容値またはイテレーションの最大回数(今回は10回)に達するまで、繰り返す  
※許容値とは、SSEをどこまで小さくするかというものの設定で、イテレーション回数が多い時に、計算コストを抑えるもの  
→SSEはクラスタが増えれば増えるほど、セントロイドとデータ間の距離が縮まるから小さくなっていく

## 3. セントロイドの初期値をスマートに設定できるk-means++法について

- kmeans法では、初期値によって、空のクラスタができるセントロイドに割り振られてしまう可能性がある
- k個のセントロイドの初期値を、セントロイド同士の距離が最も遠くなるように、設定することで解決

<b>【セントロイド同士の距離を最大化するには以下の確率分布を最大化すれば良い】</b>


$
 \frac{d( μ^{(p)} , M)^2}{\sum_{i} d(x^{(i)} , M)^2} 
$

$
μ^{(p)}：Mに含まれていないデータ点ごとにMにあるセントロイドに対して、距離の２乗(分母)が最小となるセントロイドを求める \\
x^{(i)}：データ点で、ここからセントロイドを選択していく   \\
M:セントロイドが格納されているリスト(1個目のセントロイドをランダムに格納し、２個目以降は上式の確率分布にしたがって決定され格納)
$

分子 : 2個目以降に選択されたセントロイドと最初のセントロイドの距離の2乗  
分母 : 最初のセントロイドと全ての点の距離の２乗の合計(Mに最初のセントロイドが格納されている)←これを最小にする

1. データ点の中からランダムに一つ選ぶ
2. 1のデータ点から最も近いデータ点までの距離をだす  
・ その距離が長いほど次の中心点(セントロイド)に選ばれる確率が高くなる←確率分布に依存するとこ  
・ テータ点から選ばれる中心点の候補は複数ある
3. 複数の候補の点と他のデータ点との距離の二乗の総和を計算し、総和が小さい点を中心点とする  
・ k個のセントロイドが選択されるまで2,3を繰り返す

- <b>kmeans++　を実装してみる</b>    

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
km = KMeans(
    n_clusters=2, #クラスター数
    init = "k-means++", #デフォルトはこれ ("randomにするとk個の初期値がランダムに選ばれる")
    n_init=10, #異なるセントロイドの初期値を用いたkmeansの実行回数
    max_iter=10, #kmeans内部の最大イテレーション数
    tol=1e-04, #収束と判定するためのSSEの相対的な許容範囲 
    random_state=0 #セントロイドの初期化に用いる乱数生成器の状態 0は固定
           )
```

```python
y_km = km.fit_predict(pc)
```

```python
#さすがは我らがkmeans++　元データとの比較でしっかり分類できている　
#と見せかけてシルエット係数的にはあんまり分類できていないとしている→クラスタ間の距離が欲しかった　
plt.scatter(pc[:,0],pc[:,1],c=y_km,cmap='PRGn_r')
```

## 4. ハードクラスタリングとソフトクラスタリング

<b>【クラスタリングについて】</b>

| クラスタリング                                   | 概要                                                                    | 手法                     | イメージ                     | 
| ------------------------------------------------ | ----------------------------------------------------------------------- | ------------------------ | ---------------------------- | 
| ハードクラスタリング                             | データ点がそれぞれ1つのクラスタに<br>割り当てられるアルゴリズム         | kmeans法、kmeans++法など | Aクラスに5人<br>Bクラスに3人 | 
| ソフトクラスタリング<br>(ファジークラスタリング) | データ点を1つ以上のクラスタに割り当てるアルゴリズム（クラスタがかぶる） | FCM法など                | Aクラスに70%<br>Bクラスに30% | 

- ハードクラスタリングは、データの所属関係を2値の疎ベクトル(0か1の離散値)で表現できる
- ソフトクラスタリングは、0以上1以下の連続値という該当するクラスタのセントロイドのメンバーとなる確率で表現できる  
  (クラスタの所属確率の合計は1)

 <b>【ファジー理論について簡単に】</b>  
 参考：http://www.cs.osakafu-u.ac.jp/hi/honda/fcm.htm

・YES or NO　(これらでわりきれる集合をクリスプ集合という)で割り切れないあいまいさという概念を表そうとする理論(かわいいってなに？など)    
・パターン認識が機械は弱い。。。    
・ファジーというのは、「客観的に観察された何か」ではなく主観をベースに構築された理論  

- 手法による違いのまとめ

| 手法     |メンバシップ値        |  
| -------- | --------------------- | 
| kmenas法 | 0か1           | 
| FCM法    | 0以上1以下     | 

- 数式の部分では、$w^{(i,j)}$　の部分が異なる  



<b>kmenasなどでは : </b>

$
w^{(i,j)} \in \{0,1\}
$

<b>FCMなどでは(テキスト11.1.6の式は0~1までの値を取るようにするもの) : </b>

$
w^{(i,j)} \in [0,1]
$


- 複数のクラスターへ同時に所属することを許容し，分割の様子を所属の度合いで表そうとする考えがファジーの根本
- 制約条件を0か1ではなく、0~1に幅を持たせる
- ファジーにするためには、最小にする目的関数を線形から非線形にする必要がある(最適解が0か1しか線形は取らなくなってしまうから)  
 →だから、wにべき乗(m)がつくんだ！
- mが大きいほどファジーなクラスタになる

$
min   \sum_{i=1}^{n}w^{(i,j)}{^m}\begin{Vmatrix}x^{(i)}-μ^{(j)}\\ \end{Vmatrix}^2_2
$

$
s.t. w^{(i,j)} \in \{0,1\}
$

```python
#irisデータでみてみる
# coding: UTF-8
#　パラメータmをいじると各クラスターに割り当てられる確率が変化する

import numpy as np

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans as KM

from matplotlib import pyplot as plt

#from skfuzzy.cluster import cmeans　

def target_to_color(target):
    if type(target) == np.ndarray:
        return (target[0], target[1], target[2])
    else:
        return "rgb"[target]

def plot_data(data, target, filename="fig.png"):
    plt.figure()
    plt.scatter(data[:,0], data[:,1], c=[target_to_color(t) for t in target])
    plt.savefig(filename)

def gen_data():
    iris = load_iris()
    pca = PCA(n_components=2)
    return pca.fit_transform(iris.data), iris.target

def main():
    data, target = gen_data()
    plot_data(data, target, filename="origin.png")

    km = KM(n_clusters=3)
    km_target = km.fit_predict(data)
    plot_data(data, km_target, filename="kmeans.png")

    for m in [1.5, 2, 3, 4, 5]:
        cm_result = cmeans(data.T, 3, m, 0.003, 10000)
        plot_data(data, cm_result[1].T, filename="cmeans_{0}.png".format(m))

if __name__ == "__main__":
    main()
```

## 5. エルボー法を用いたクラスタの最適な個数

- クラスタリングに最適なクラスタの個数がわからない
- kmenasは、あるクラスタ数に対するSSE(クラスタ内誤差平方和)を最小にするのが目的
- エルボー法は、逐次的にクラスタリングを試してSSEが急激に下がる(これ以上増やしてもSSEの減少に大きく寄与しない)  
  クラスタ数を発見することが目的

```python
distortions = []
for i in range(1,11):
    km_d = KMeans(
        n_clusters=i, #クラスター数
        init = "k-means++", #デフォルトはこれ ("randomにするとk個の初期値がランダムに選ばれる")
        n_init=10, #異なるセントロイドの初期値を用いたkmeansの実行回数
        max_iter=10, #kmeans内部の最大イテレーション数
        tol=1e-04, #収束と判定するためのSSEの相対的な許容範囲 
        random_state=0 #セントロイドの初期化に用いる乱数生成器の状態 0は固定
           )
    km_d.fit(pc)
    distortions.append(km_d.inertia_)

plt.plot(range(1,11),distortions,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

k = 2 がこのデータセットでは良いとされる

この図を見たら、10分割することが一番SSEを小さくするが、当たり前だが重心(セントロイド)の数が増えれば、各点との距離は縮まる。。。

```python
km_10 = KMeans(
    n_clusters=10, #クラスター数
    init = "k-means++", #デフォルトはこれ ("randomにするとk個の初期値がランダムに選ばれる")
    n_init=10, #異なるセントロイドの初期値を用いたkmeansの実行回数
    max_iter=10, #kmeans内部の最大イテレーション数
    tol=1e-04, #収束と判定するためのSSEの相対的な許容範囲 
    random_state=0 #セントロイドの初期化に用いる乱数生成器の状態 0は固定
           )
y_km_10 = km_10.fit_predict(pc)
```

## 6. シルエット図でクラスタリングの性能を数値化する

1. データ点がどの程度密にグループ化されているかという凝集度の計算(同一クラスタのデータ点と他の全データ点との平均距離として計算)  
   →同じクラスタ内のデータの密集度
2. 最も近いクラスタからの乖離度を計算(データ点と最も近くにあるクラスタ内の全データ点との平均距離で計算)  
   →クラスタ同士のデータの離れ具合
3. クラスタの凝集度と乖離度の差を、そのうちの大きい方の値でわり、シルエット係数$s^{(i)}$を計算する

$
s^{(i)} =\frac{b^{(i)} - a^{(i)}}{ max\{b^{(i)},a{(i)}\}}
$

$
b^{(i)}：乖離度  \\ 
a^{(i)}：凝集度　　 \\
$

- シルエット係数は-1から1
- クラスタの乖離度と凝集度が等しい時、シルエット係数は0になる  
- $b^{(i)} >> a^{(i)}$  の時、シルエット係数は1に近づく

```python
def silet(X,n):
    from sklearn.cluster import KMeans
    km = KMeans(
        n_clusters=n, #クラスター数
        init = "k-means++", #デフォルトはこれ ("randomにするとk個の初期値がランダムに選ばれる")
        n_init=10, #異なるセントロイドの初期値を用いたkmeansの実行回数
        max_iter=10, #kmeans内部の最大イテレーション数
        tol=1e-04, #収束と判定するためのSSEの相対的な許容範囲 
        random_state=0 #セントロイドの初期化に用いる乱数生成器の状態 0は固定
           )
    km.fit(X)
    pred = km.predict(X)
    from yellowbrick.cluster import SilhouetteVisualizer
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick')
    visualizer.fit(X)
    visualizer.show()
```

```python
#c = 2のときのシルエット図
silet(pc,2)
```

```python
#c=10の時のとき 
silet(pc,10)
```

```python
n_clusters = [2, 4, 6, 10]
colors = ['lightgreen', 'orange', 'lightblue', 'hotpink', 'yellow','red','Cyan','Magenta','tomato','linen']
markers = ['s', 'o','v', 'h','p','<','>','*','x','1']

for plot_num, n_cluster in enumerate(n_clusters):

x_data = pc

# k-meansは一行で行える
# n_clustersでクラスタ数を指定し、initで初期化方法を指定している
# initの初期化は「k-means++」であり、これは後ほど説明する
km = KMeans(n_clusters=n_cluster, init='k-means++')

predicted_labels = km.fit_predict(x_data)
centroids = km.cluster_centers_

# クラスタリング結果の可視化
plt.subplot(2, 2, plot_num+1)
for i, label in enumerate(sorted(np.unique(predicted_labels))):

    # 分類した結果のクラスをプロット
    plt.scatter(x_data[predicted_labels==label, 0],
                x_data[predicted_labels==label, 1],
                c=colors[i],
                marker=markers[i],
                s=15)

# 最終的なセントロイドをプロット
plt.scatter(centroids[:, 0],
            centroids[:, 1],
            c='black',
            marker='*',
            s=70)

plt.title('n cluster '+str(n_cluster))

plt.tight_layout()
plt.show()
```
