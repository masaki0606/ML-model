## ■ (フィッシャーの)線形判別分析（LDA）は次元削減をして、データの圧縮を行うもの

1. 次元削減は記憶域や計算効率の改善の他に次元の呪いを減らすことで予測性能を向上させることが目的
 
 
2.  次元削減には特徴量選択と特徴量抽出がある　　
 - 特徴量選択：特徴量の一部を選択　(ex 逐次特徴量選択アルゴリズムなど)
 - 特徴量抽出：特徴量の集合から情報を抽出　　(ex PCA,LDAなど)

## ■ LDAは、特徴量抽出という手法に該当

- 特徴量抽出とは、、、  

1.  正則化されていない（多重共線性のある）「次元の呪い」による過学習を抑制するために利用

 - 正則化：モデルが複雑にならないように、モデルを学習する際に複雑さが増すことに対するペナルティを設けて過学習を防ぐ  
     L1正則化：特定のデータの重みを0することで、不要なデータを削除する(係数の絶対値の和)  
     L2正則化：データの大きさを０に近づけて、滑らかなモデルとする(係数の二乗和)
 - 次元の呪い：次元の数が増えれば増えるほど、一般化するために必要な訓練データの量が「指数関数」的に増えてしまう   
    →特徴量どうしの組み合わせ問題（めっちゃ計算コストが増える、無関係なはずの特徴量にまで意味を見出してしまうことも）
 - 過学習：訓練データにフィットしすぎて、未知のデータに対応できなくなること
 
    
2. クラスの分離を最適化する特徴量部分空間を見つけることが目的    
    →特徴量ベクトルのN次元空間をうまくグルーピングする

## ■ 判別分析はデータの群わけを最適に判断する方法(判別分析のイメージ)

- 多変量データを持つ個体をどう判別すればいいのか。。。  
  例えば、肥満体な人間を表すために、体重が１００kgであればいいか？？といえば、違う。身長も踏まえて、BMIという指数で判別する。  
 　  →身長、体重という２変量から、BMIという１変量に判別基準を減らすことができた！　

## ■ LDAは相関比を最大にすること

- 相関比
$$J(w)=　クラス間変動　/　クラス内変動 $$
- 線形判別関数(相関比を最大にするように係数と切片を設定、例えば2クラス)
$$ \hat{y} = w_1x_1 + w_2x_2 + w_0$$   
※相関比を用いて、係数を決定する。切片は２群の平均の中点が原点に来るように決定(2変数の式)。



- 線形判別分析の導出
   カテゴリごとの平均からクラス間変動を、カテゴリごとの値とカテゴリごとの平均からクラス内変動を求めると以下のようになる。 
  
$$J(w) = W^TS_BW/W^TS_WW$$
  
   上記は、二次形式の式なので、実数値になっている。これを最大化するW(1×dのベクトル)を求めるのが目的。  
   ただし、分母を0にすると∞に発散するので、制約条件が必要となる。
   
   
   
- 制約条件　
  $$W^TS_WW=1$$


- 線形判別分析とは、制約条件ありの相関比を最大化してWを求める問題  
  $$
   max_w W^TS_BW
  $$
  $$
   s.t. W^TS_WW = 1
  $$

- 上記の制約条件付きの相関比最大化問題をとくと
  
  $$
   S_W^-1S_B・W = λW
  $$
  
  の固有値問題に置き換えることができる。


1. 仮説(入力xに対して)：  

  - 2クラス分類  　
  $$
    \mathrm y_i =
        \begin{cases}
            1 \quad  \hat{y} > 0 \\ 
            -1 \quad \hat{y}\leqq 0 \\
        \end{cases}
$$
  - 多クラス分類  
  $$
     y_i ∈ Y = {1, 2, . . . , G}
  $$
  
2. 損失関数(最大値を求めるけど、多クラスでも同じ結果になる)：  
　$$J(w) = S_W^-1 S_B$$

固有値の大きいものに対応する固有値ベクトルほど、射影された特徴量データはクラス間変動が大きくクラス内変動が小さくなる。

## ■ 主成分分析との違い

主成分分析は、分散が最大になるような合成変量を説明変数のデータから新しく作り、次元削減をするのに対して、線形判別分析は目的変数のデータを射影して相関比が最大となるように次元削減を行う。
だから、教師ありデータの次元削減になる。

## ■ 線形判別分析の内部の仕組み

1. d次元のデータセットを標準化する（N(0,1)）。データの大きさに分散が依存してしまうのを防ぐ。
2. クラスごとにd次元の平均ベクトル(各次元の平均値で構成されるベクトル)を計算する。
3. 平均ベクトルを使って、クラス間(群間)変動行列Sb　と、クラス内(郡内)変動行列Swを生成する。　
4. 行列Sw-1Sbの固有値ベクトルと対応する固有値を計算する。
5. 固有値を降順でソートすることで、対応する固有ベクトルをランクつけする。
6. d × k 次元の変換行列Wを生成するために、最も大きいk個の固有値に対応するk個の固有ベクトルを選択する（固有ベクトルから変換行列Wを生成）。  
   固有ベクトルは、この行列の列である。
7. 変換行列Wを使って、データを新しい特徴量部分空間へ射影する。

## ■ Wineのデータセットを用いて、線形判別分析を行う

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/''machine-learning-databases/wine/wine.data',header=None)

df_wine.columns = ['Class label','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium',
                   'Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity',
                   'Hue','OD280/OD315 of diluted wines','Prolines'] 
##クラスラベルが目的変数になる
```

```python
df_wine
```

### ▼ STEP_1 :  d次元のデータセットをN(0,1)になるように標準化する

```python
from sklearn.model_selection import train_test_split ##テキストにある cross_validationは廃止
from sklearn.preprocessing import StandardScaler
X,y = df_wine.iloc[:,1:].values,df_wine.iloc[:,0].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=0) ##訓練データとテストデータに分割
sc =StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
##訓練データを学習（変換式を作る）と値を変換
##テストデータを変換（訓練データで作られた変換式で変換する）
```

```python
X_train_std
```

```python
y_train
```

### ▼STEP_2 : クラスごとにd次元の平均ベクトル(各次元の平均値で構成されるベクトル)を計算

```python
np.set_printoptions(precision=4) ##表示桁数を指定　小数点以下４桁まで設定
mean_vecs = []
for label in range(1,4):     
    mean_vecs.append(np.mean(X_train_std[y_train==label],axis=0))
    print('MV %s: %s \n' %(label,mean_vecs[label-1])) 
    ##目的関数のクラスラベルの数(1~3)だけ平均値ベクトルを作成（クラスごとの平均）
    ##例えば、クラス1のAlcoholの平均が0.9066のようになる
```

### ▼STEP_3：平均ベクトルを使って、クラス内(郡内)変動行列Swとクラス間(群間)変動行列Sbを生成

・クラス内変動行列 ：　個々のデータと個々のクラスの平均との差を用いた平方和

```python
d = 13 ##特徴量数
S_W = np.zeros((d,d)) ##13 × 13の零行列を作成
for label , mv in zip(range(1,4),mean_vecs):    ##zip関数で1~3のクラスラベルと平均ベクトルを変数であるlabelとmvに格納
    class_scatter = np.zeros((d,d))
    for row in X_train_std[y_train == label]:  ##クラスラベルに対応するtrainデータをrowに格納 →式のxにあたる
        row,mv = row.reshape(d,1),mv.reshape(d,1) ##(x-mi)に関わる部分をrowとmvで作る
        class_scatter += (row - mv).dot((row - mv).T) ##(x-mi)(x-mi)^Tを計算
    S_W += class_scatter
print('With-class scatter matrix: %sx%s' % (S_W.shape[0],S_W.shape[1]))
```

```python
mean_vecs ##平均ベクトル
```

```python
S_W
```

- クラスラベル(y_train)が一様に分布している必要があるが、全然していないので、スケーリングが必要。

```python
print('Class label distribution : %s' %np.bincount(y_train)[1:])
```

- スケーリングでは、変動行列の要素をそれぞれのクラスラベルの数で割る。

```python
d = 13
S_W = np.zeros((d,d))
for label , mv in zip(range(1,4),mean_vecs):
    class_scatter = np.cov(X_train_std[y_train == label].T)##共分散行列を求める
    S_W += class_scatter
print('Scaled with-class scatter matrix :  %sx%s' % (S_W.shape[0],S_W.shape[1]))
```

```python
S_W
```

・　クラス間変動（クラス間共分散行列）を求める  
　　→個々のクラスの平均と全体平均の差を用いた平方和に個々のクラスにある数を重みとして乗じたもの（スケーリングは必要ない）

```python
mean_overall = np.mean(X_train_std,axis=0) ##行方向の平均
d = 13
S_B = np.zeros((d,d))
for i ,mean_vecs in enumerate(mean_vecs):
    n = X_train_std[y_train == i + 1,:].shape[0] ##i==0からスタートしているから+1 、クラスごとの行数を取得(重みのため)
    mean_vecs = mean_vecs.reshape(d,1) ##13行1列の列ベクトルを作成
    mean_overall = mean_overall.reshape(d,1) 
    S_B += n * (mean_vecs - mean_overall).dot((mean_vecs - mean_overall).T) 
```

```python
S_B
```

```python
print(('Between-class scatter matrix: %sx%s' % (S_B.shape[0],S_B.shape[1])))
```

### ▼STEP_4 : 新しい特徴量部分空間の線形判別を選択する

- 特徴量部分空間：ベクトル空間において、ベクトル空間の演算によって閉じている部分集合が部分空間。    
  部分空間の要素は、ベクトルの和とスカラー倍で座標を表すことができる。     

```python
eigen_vals,eigen_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B)) ##固有値を求める inv：逆行列　dot：行列積 eig：関数の固有値を計算
```

```python
eigen_pairs = [(np.abs(eigen_vals[i]),eigen_vecs[:,i]) ##eigen_vals(固有値)の絶対値を取り、固有ベクトルと固有対にする
                         for i in range(len(eigen_vals))]
eigen_pairs = sorted(eigen_pairs,key=lambda k: k[0],reverse=True) ##無名関数lambda 要素kを受け取りk[0]を返す(k[0]を基準に並べる)
print('Eigenvalues in descending order: \n')
for eigen_val in eigen_pairs:
    print(eigen_val[0])
```

```python
eigen_vals
```

```python
eigen_vecs
```

```python
##今回の固有値は、クラス間変動を最大にし、クラス内変動を最小にする変換行列を求めるための固有値
tot = sum(eigen_vals.real) ##固有値の実数部分の総和 （=共分散行列の対角である分散の総和）
discr = [(i/tot) for i in sorted(eigen_vals.real , reverse=True)] ##それぞれの固有値の実数部分 / 固有値の実数部分の総和  ←寄与率に当たる
cum_discr = np.cumsum(discr)  ##discrの累積和
plt.bar(range(1,14),discr,alpha=0.5,align='center', 
       label = 'Individual "discriminability" ')
plt.step(range(1,14),cum_discr,where='mid',
        label='Cumulative "discriminability" ')
plt.ylabel = (' "Discriminability" ratio')
plt.xlabel = ('Linear Discriminants')
plt.ylim([-0.1,1.1])
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
##上位2つの固有ベクトルから変換行列を作成
w = np.hstack((eigen_pairs[0][1][:,np.newaxis].real, ##配列ndarrayに次元を追加する newaxis 
              eigen_pairs[1][1][:,np.newaxis].real)) ##hstackは横結合 ##降順に並んだので、頭からふたつが大きい固有値にあたる
print('Matrix W:\n',w)
print(w.shape)
```

```python
eigen_pairs[0][1][:,np.newaxis].real
```

```python
X_train_Ida = X_train_std.dot(w) ##標準化したデータに変換行列をかける
X_train_std
```

```python
print(np.unique(y_train))
print(X_train_Ida[y_train == 2,0])
print(X_train_Ida[y_train == 2,1])
```

```python
colors = ['r','b','g']
markers = ['s','x','o']
for l,c,m in zip(np.unique(y_train),colors,markers):
    plt.scatter(X_train_Ida[y_train == l,0],X_train_Ida[y_train == l,1]*(-1),
               c=c,label=l,marker=m)
plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

### ▼ライブラリを用いた線形判別分析

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/''machine-learning-databases/wine/wine.data',header=None)

df_wine.columns = ['Class label','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium',
                   'Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity',
                   'Hue','OD280/OD315 of diluted wines','Prolines'] 
##クラスラベルが目的変数になる
```

```python
from sklearn.model_selection import train_test_split ##テキストにある cross_validationは廃止
from sklearn.preprocessing import StandardScaler
X,y = df_wine.iloc[:,1:].values,df_wine.iloc[:,0].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=0) ##訓練データとテストデータに分割
sc =StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
##訓練データを学習（変換式を作る）と値を変換
##テストデータを変換（訓練データで作られた変換式で変換する）
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
X_train_lda = lda.fit_transform(X_train_std,y_train)
```

```python
##ロジスティック回帰の分類器が、LDAによって変換された低次元の訓練データセットをどのように処理するか
##訓練データ
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_decision_regions
lr = LogisticRegression(multi_class='ovr',random_state=1,solver='lbfgs')
lr = lr.fit(X_train_lda,y_train)
```

```python
##訓練データ
plot_decision_regions(X_train_lda,y_train,clf=lr)
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
##sklearn の0.24.1ではまたちょっとずれた
##第1新変量(LD1) 第2新変量(LD2)
```

```python
#テストデータ
X_test_lda = lda.transform(X_test_std)
plot_decision_regions(X_test_lda,y_test,clf=lr)
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
```

2つの新変量(固有ベクトル)が残るので、１３次元あったものを２次元に圧縮できたことを示す。
